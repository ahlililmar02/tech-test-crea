```python
import json
import geopandas as gpd

# Load the GeoJSON file
with open(r'C:\Users\ahlil\Tes\stations.geojson', 'r') as f:
    stations_geojson = json.load(f)

# Convert stations GeoJSON to GeoDataFrame
stations_gdf = gpd.GeoDataFrame.from_features(stations_geojson["features"])

stations_gdf
```

```python
file_path2 = "~/Tes/countries.geojson"
countries_gdf = gpd.read_file(file_path2)
countries_gdf
```

```python
# Filter countries of interest
target_countries = ["US", "GB", "TR", "TH", "PH", "IN"]
countries_gdf_t = countries_gdf[countries_gdf["ISO_A2"].isin(target_countries)]
```

```python
countries_gdf_t
```

```python
# Calculate area in square kilometers
countries_gdf_t["Area_km2"] = countries_gdf_t["geometry"].to_crs(epsg=3395).area / 1e6

# Count PM10 stations per country
# Filter stations with 'pm10' in the 'pollutants' column
stations_gdf_pm10 = stations_gdf[stations_gdf["pollutants"].apply(lambda x: x is not None and 'pm10' in x)]

# Count PM10 stations per country
stations_count = stations_gdf_pm10.groupby("country_id").size().reset_index(name="PM10_Stations")
stations_count
```

```python
merged_gdf = countries_gdf_t.merge(stations_count, left_on="ISO_A2", right_on="country_id", how="left")
```

```python
# Calculate density
merged_gdf["Density"] = merged_gdf["PM10_Stations"] / merged_gdf["Area_km2"] * 1000
```

```python
# Select only the relevant columns
result = merged_gdf[['ADMIN', 'country_id', 'Area_km2', 'PM10_Stations', 'Density']]
result= result.sort_values(by="Density", ascending=False)
result = result.rename(columns={
    'ADMIN': 'Country Name',
    'country_id': 'Country ID',
    'Area_km2': 'Area (km^2)',
    'PM10_Stations': 'PM10 Stations',
    'Density': 'Density'
})
result
```

```python
# Save the selected table to a CSV file
result.to_csv('result.csv', index=False)
```
